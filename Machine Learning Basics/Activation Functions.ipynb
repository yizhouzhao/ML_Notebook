# Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt

def plot(func):
    x = np.arange(-10, 10)
    y = list(map(func, x))
    plt.title(str(func))
    plt.plot(x, y)
    plt.show()
```

## Sigmoid

### Equations

$$\sigma(x) = \frac{1}{1 + e^{-x}}$$


```python
def sigmoid(x):
    return 1./(1. + np.exp(-1*x))
```

```python
plot(sigmoid)
```

### Properties

1. Outputs value between $0$ and $1$
2. Control Signal in LSTM network (?)
3. Used for Binary functions
4. Not for image networks (WHY?)
5. Not Zero Centered, (What is the problem with this feature?)

**What happens if input to a neuron is always positive?**

## Tanh

### Equations

$$\tanh(x) = \frac{\sinh{x}}{\cosh(x)} = \frac{e^z - e^{-z}}{e^z + e^{-z}} = \frac{e^{2z} - 1}{e^{2z} + 1}$$

```python
def tanh(x):
    return np.tanh(x)

plot(tanh)
```

### Properties

1. Range between $-1$ and $1$
2. Zero Centered
3. Gradient Saturation :(
4. Used in LSTM for bounded signed values
5. Not for binary functions

## ReLU

### Equations

$$f(x) = \max(0, x)$$

```python
def relu(x):
    return max(x, 0)

plot(relu)
```

### Properties

1. Does not saturate
2. Converges faster than sigmoid/tanh on image data (6x)
3. Not suitable for logical functions
4. Not for control in recurrent network
5. Not zero centered
6. What if the value is negative?  (dead RELU)

## Leaky ReLU

### Equations

$$f(x) = \max(0.01x, x)$$

```python
def leaky_relu(x):
    return max(0.05*x, x)

plot(leaky_relu)
```

### Properties
1. Does not saturate
2. Does not die
3. Converges faster than sigmoid/tanh on image data

**Note**: $0.05$ can be parametrized- **Parameter Rectifier (PReLU)** $f(x) = \max(\alpha x, x)$

## ELU

### Equations

$$f(x) = \begin{cases}
               x & \text{if $x  > 0$} \\
               \alpha (\exp(x) - 1) & \text{if $x \leq 0$} \\
          \end{cases}$$

```python
def elu(x, alpha=0.1):
    if x > 0:
        return x
    return alpha*(np.exp(x) - 1)
plot(elu)
```

### Properties
1. All benefits of ReLU
2. Doesn't die
3. Closer to zero mean outputs

## Maxout Neurons

### Equations

$$f(x, w, b) = max(w_1^Tx + b, w_2^Tx + b_2)$$

### Properties
1. Linear Regime
2. Doesn't saturate or die
3. Generalizes ReLU and Leaky ReLU
4. Too many parameters
